```python
import pandas as pd
import numpy as np
```

```python
#Read Galaxy file
df = pd.read_csv("galaxy2.csv")
```

```python
df
```

```python
del df['length']
```

```python
del df['eff_length']
del df['est_counts']
```

```python
#Read best match file
df2 = pd.read_csv("whiteselli_best.csv")
```

```python
df2
```

```python
#Change name of 'sequence_id' column to 'target_id'
df2.rename(columns={'sequence_id':'target_id',
                    }, 
                 inplace=True)
```

```python
#eye pigmentation
df2_1 = df2[:-52]
```

```python
df2_1
```

```python
#Add tpm to eye gene
df3_1 = pd.merge(df2_1, df, on=['target_id'], how='inner')
```

```python
df3_1
```

```python
#phototransduction
df2_2 = df2[28:61]
```

```python
df2_2
```

```python
#Add tpm to phototransduction gene
df3_2 = pd.merge(df2_2, df, on=['target_id'], how='inner')
```

```python
df3_2
```

```python
#circadian rhythm gene
df2_3 = df2[63:]
```

```python
df2_3
```

```python
#Add tpm to circadian gene
df3_3 = pd.merge(df2_3, df, on=['target_id'], how='inner')
```

```python
df3_3
```

```python
#photolyase
df2_4 = df2[62:63]
```

```python
df2_4
```

```python
#Add tmp to photolyase
df3_4 = pd.merge(df2_4, df, on=['target_id'], how='inner')
```

```python
#Add phototransduction row
data1 = df3_1.append({'query': 'phototransduction'}, ignore_index=True)
```

```python
#Add circadian rhythm row
data2 = df3_2.append({'query': 'circadian rhythm gene'}, ignore_index=True)
```

```python
#Add photolytase row
data3 = df3_3.append({'query': 'photolytase'}, ignore_index=True)
```

```python
#Create final table
table = pd.concat([data1, data2, data3, df3_4])
```

```python
table
```

```python
#Export final file
table.to_csv("best_match_with_tpm.csv",index = None, header=True)
```
